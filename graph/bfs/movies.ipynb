```python
class Node:
    def __init__(self, value):
        self.value = value
        self.edges = []
        self.visited = False
        self.parent = None

    def add_edge(self, other):

        self.edges.append(other)
        other.edges.append(self)

    def reset(self):
        self.visited = False
        self.parent = None

    def __repr__(self):
        return f"value = ({self.value}), visited = {self.visited}, parent = {self.parent.value if self.parent is not None else None}"
    
    def __str__(self):
        return f"{self.value}"
```

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_node(self, node):
        # if not isinstance(node, Node):
        node = Node(node)
        value = node.value
        # if not value in self.graph:
        self.graph[value] = node
        return node

    def get_node(self, node):
        return self.graph.get(node, None)

    def __repr__(self):
        return str(len(self.graph))

    def reset(self):
        for node in self.graph:
            self.get_node(node).reset()
```

```python
graph = Graph()
```

```python
for line in data:
    elms = line.split('/')
    movie = elms[0]
    actors = elms[1:]
    movie_node = graph.add_node(movie)
    for actor in actors:
        actor = actor.strip()
        actor_node = graph.get_node(actor)
        if actor_node == None:
            actor_node = graph.add_node(actor)
        actor_node.add_edge(movie_node)
```

```python
from collections import deque
```

```python
def bfs(graph, source="Keanu Reeves", dest="Kevin Bacon"):
    graph.reset()

    queue = deque([graph.graph[source]])
    while len(queue) > 0:
        node = queue.popleft()
        node.visited = True
        for edge in node.edges:
            if edge.visited == False:
                queue.append(edge)
                edge.visited = True
                edge.parent = node
                if edge.value == dest:
                    print("Found!!")
                    path = []
                    while edge:
                        path.append(edge)
                        temp = edge
                        edge = edge.parent
                        temp.reset()
                    return path
```

```python
for node in bfs(graph):
    print(node)
```
